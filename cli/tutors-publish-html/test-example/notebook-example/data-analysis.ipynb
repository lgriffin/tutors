```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set up matplotlib for inline plotting
plt.rcParams['figure.figsize'] = (10, 6)

print("Libraries imported successfully!")
```

```python
# Generate sample data
np.random.seed(42)
n_samples = 100

data = {
    'age': np.random.randint(18, 65, n_samples),
    'income': np.random.normal(50000, 15000, n_samples),
    'satisfaction': np.random.randint(1, 11, n_samples)
}

df = pd.DataFrame(data)
df['income'] = df['income'].clip(lower=20000)  # Ensure no negative income

print(f"Dataset created with {len(df)} rows and {len(df.columns)} columns")
print("\nFirst 5 rows:")
df.head()
```

```python
# Create a simple visualization
plt.figure(figsize=(12, 4))

# Age distribution
plt.subplot(1, 2, 1)
plt.hist(df['age'], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Income vs Satisfaction scatter plot
plt.subplot(1, 2, 2)
plt.scatter(df['income'], df['satisfaction'], alpha=0.6, color='purple')
plt.title('Satisfaction vs Income')
plt.xlabel('Income ($)')
plt.ylabel('Satisfaction Score')

plt.tight_layout()
plt.show()
```
